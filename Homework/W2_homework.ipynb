## Exersice 22

### Answers:
* The expected value of the price of the stock after d days is $\mathbb{E}[q_d]=r^{d(2p-1)}$
* The variance of the price of the stock after d days is $\mathrm{Var}(q_d)=\left(p\cdot r^2+(1-p)\cdot r^{-2}\right)^d-r^{2d(2p-1)}$


### Solution:

#### 1. Expected value

According to the question, the price of stock in $n$th day is:

$$
    q_n=\begin{cases}q_n\cdot r,p\\q_n/r,1-p\end{cases}
$$

We can simplify as:

$$
    q_n = q_{n-1}\cdot r^{X_i}
$$

where:

$$
    X_i=\begin{cases}1&\text{with probability }p\\-1&\text{with probability }1-p\end{cases}
$$

The price of stock over $d$ days

$$
    q_d=1\cdot r^{X_1}\cdot r^{X_2}\cdot\cdots\cdot r^{X_d}=r^{\sum_{i=1}^dX_i}
$$

Thus the expected price is:

$$
    E[q_{d}]=E[r^{\sum_{i=1}^d X_i}]=r^{E[\sum_{i=1}^d X_i]}=r^{\sum_{i=1}^d E[X_{i}]}
$$

Given that:

$$
    E[X_i] = 1\cdot p + (-1)\cdot (1-p) = 2p-1
$$

$$
   E[q_d]=r^{d(2p-1)}
$$

#### 2. Variance of price

To calculate the variance, we need to calculate $E[{q_d}^2]$ first

$$
    {q_d}^2=r^{2(X_1+X_2+\cdots+X_d)}=r^{\sum_{i=1}^{d}2X_i}
$$

Because $X$ is independent, thus:

$$
    E[{q_d}^2]=\prod_{i=1}^{d}E[r^{2X_{i}}]
$$

For $X_i$:

$$
    E[r^{2X_i}]=p\cdot r^2+(1-p)\cdot r^{-2}
$$

Thus:

$$
    {E}[{q_d}^2]=\left(p\cdot r^2+(1-p)\cdot r^{-2}\right)^d
$$

As a result:

$$
    Var(q_d)=E[q_d^2]-(E[q_d])^2=\left(p\cdot r^2+(1-p)\cdot r^{-2}\right)^d-r^{2d(2p-1)}
$$

```python
# Python part for exercise 6
import numpy as np
import matplotlib.pyplot as plt

# Parameters
r = 1.1
p = 0.5
q0 = 1
day = 50
num = 200

# Empirical
np.random.seed(42)
price_list = np.zeros((num, 51))
price_list[:, 0] = q0

for i in range(1, 51):
    rate = np.random.choice([r, 1/r], size=num, p=[p, 1 - p])
    price_list[:, i] = price_list[:, i - 1] * rate

# Estimate expected value and variance for each day
expected_prices = np.mean(price_list, axis=0)
variance_prices = np.var(price_list, axis=0)

# Theoretical
d = np.arange(51)
theoretical_expectation = q0 * r**(d * (2 * p - 1))
theoretical_variance = (q0**2) * ((p * r**2 + (1 - p) * (1/r)**2)**d - (r**(2 * d * (2 * p - 1))))


plt.figure(figsize=(10, 5))

# Plot expected value
plt.subplot(1, 2, 1)
plt.plot(d, expected_prices, label="Simulated Expected Price")
plt.plot(d, theoretical_expectation, label="Theoretical Expected Price", linestyle='--')
plt.xlabel('Days')
plt.ylabel('Expected Price')
plt.title('Expectation over Time')
plt.legend()

# Plot variance
plt.subplot(1, 2, 2)
plt.plot(d, variance_prices, label="Simulated Variance")
plt.plot(d, theoretical_variance, label="Theoretical Variance", linestyle='--')
plt.xlabel('Days')
plt.ylabel('Variance')
plt.title('Variance over Time')
plt.legend()

plt.show()
```

## Exersice  27

### Answers: 
* The best upper bounds on p using Markov’s inequality is $\frac23$
* The best upper bounds on p using Chebyshev’s inequality is $\frac{20}{n}$
* The best upper bounds on p using Chernoff bounds is $\exp\left(-\frac{n}{72}\right)$

### Solution:

According to question, we can conclude:

* $X\sim\mathrm{Binomial}(n,\frac16)$
* $E[X]=\frac{n}{6}$
* $Var[x]=E[x^{2}]-E[x]^{2}=\frac{5}{36}n$

#### Markov’s inequality

Definition of Markov's inequality:

$$
    P(X\geq a)\leq\frac{E[X]}a
$$

Set $a$ as $\frac{n}{4}$:

$$
    \mathbb{P}(X\geq\frac n4)\leq\frac{\frac n6}{\frac n4}=\frac23
$$

#### Chebyshev’s inequality

Chebyshev's inequality is:

$$
    P[x-E[x]\geq a]\leq\frac{Var[x]}{a^{2}}
$$

Given that $E[X]=\frac{n}{6}$, we can set $a=\frac{n}{12}$, then:

$$
    P[x\geq\frac{3n}{4}]=P[x-E[x]\geq\frac{n}{12}]\leq\frac{\frac{5n}{36}}{(\frac{n}{12})^{2}}=\frac{20}{n}
$$

#### Chernoff bounds

To use Chernoff bounds, we need to calculate $\delta$ first.

Because

$$
    \begin{cases}u=E[x]=\frac n6\\\\(1+\delta)u=\frac n4\end{cases}
$$

So:

$$
    \delta = \frac{1}{2}
$$

For every $0<\delta\leq1$:

$$
    \mathbb{P}[X\geq(1+\delta)\mu]\leq e^{-\mu\delta^2/3}
$$

Sustitute $\delta = \frac{1}{2}$, $u=\frac14$:

$$
    \begin{aligned}P[x\geq\frac{n}{4}]=P[x\geq(1+8)n]&\leq\exp(-u\cdot\frac{\delta^{2}}{3})\\&=\exp(-\frac{n}{72})\end{aligned}
$$

```python
# Placeholder for python part of exersice 27
import numpy as np
import matplotlib.pyplot as plt

# Parameters
p = 1/6
n_list = np.arange(100, 151)
instances = 1000

x1_list = np.zeros((instances, len(n_list)))
p_list = np.zeros(len(n_list))

# Empirical
for n_i in range(len(n_list)):
    for i_m in range(1,1001):
        x_list = np.random.choice([1, 0], replace=True, size=n_list[n_i], p=[p, 1 - p])
        x1_list[i_m-1, n_i] = np.sum(x_list == 1)

for p_i in range(len(p_list)):
    p_list[p_i] = np.mean(x1_list[:, p_i] >= n_list[p_i] / 4)

# Theoretical
Markov_b = np.full((51, 1), 2/3)
Chebyshev_b = 20 / n_list
Chernoff_b = np.exp(- n_list / 72)


plt.figure(figsize=(15, 6))
plt.plot(n_list, p_list, label="Estimated P", marker='o', linestyle='--', color='blue')
plt.plot(n_list, Markov_b, label="Markov_b",  color='orange')
plt.plot(n_list, Chebyshev_b, label="Chebyshev_b", color='yellow')
plt.plot(n_list, Chernoff_b, label="Chernoff_b", linestyle='--', color='red')
plt.xlabel('Throw Number of Dice(n)')
plt.ylabel('Probability of The Event X ≥ n/4')
plt.title('Campare of Different Bounds ')
plt.legend()

plt.show()
```

##### Conclusion
The chart inllustrates that:

1. When $n$ is small：
    - Chebyshev’s inequality gives the best approximation.
        
2. When $n$ is large and Towards infinity：
    - Chernoff bounds gives the best approximation.
    - It can be estimated that as n goes to infinity, Chernoff bounds gives a better and better approximation.
