## Exersice 06 (Virus Test)
A medical company is assessing a newly developed
test for a certain virus. The so called false negative rate is small: if you have
contracted the virus, the probability that the test returns a positive result is
0.999. The false positive rate is also small: if you do not have the virus, the
probability that the test returns a positive result is only 0.005. Assume that 2%
of the population has contracted the virus. If a person chosen uniformly from the population is tested and the result comes back positive, what is the probability that the person has the virus?
### Answers:
Fill in your anwers here so that Carl can do a quick check if they are right/wrong: 

The probability that the person has the virus is $\approx$ 80.3%



### Solution:


Lets define the following parameters:

- $V$: People contracted the virus

- $T_p$:  The medical test result is positve

- $T_{fp}$: The medical test result is false positive

- $P[V]$ = 0.02 as(Assumed 2% of the population has contracted the virus)

- $P[V_c]$ = 0.98 as(Assumed 98% of the population has not contracted the virus)

- $P[T_p| V]$ = 0.999 (probability that the test returns a positive result for person has contracted the virus)

- $P[T_{fp} | Vc]$ = 0.005 (probability that the test returns a false positive result for person has not contracted the virus)

Required: what is the probability that the person has the virus? ---> $P(V | T_p)$

$$
\Pr[V | T_p] = \frac{\Pr[T_p | V] * \Pr[V]} {\Pr[T_p]}
$$

$$
\Pr[V | T_p] = \frac{0.999 * 0.02} {\Pr[T_p]}
$$

Using the law of total probability:

$$
\Pr[T_p] = \Pr[T_p | V] * \Pr[V] + \Pr[T_{fp} | V_c] * \Pr[V_c]
$$

$$
\Pr[T_p] = 0.999 * 0.02 + 0.005 * 0.98 = 0.02488 
$$

So 
$$
\Pr[V | T_p] = \frac{0.999 * 0.02} {0.02488 } = 0.803054662
$$


```python
# Placeholder for python part of exersice 06
# python: Draw N = 10^4 pseudo random samples representing the infection
# status of N individuals. Use these samples and the statistics described above for
# the virus test to draw pseudo random samples representing a positive or negative
# virus test for each individual and verify your results from above.

import numpy as np
import matplotlib.pyplot as plt

def virus_test(no_individuals):

    # Generate test results with false
    test_results = np.zeros(no_individuals, dtype=bool)
    # Generate sample of expected outcomes with probabilty 2% of the individauls has contracted the virus
    # the below function will generate an array with values to choose from [True, False] which means [infected, not infected]
    # ,with array size equal to N and the probabilty is 2% infected indvividuals with virus.
    people_sample = np.random.choice([True, False], size=no_individuals, p=[0.02, 0.98])
    # fill array of test_results with test value ([Postive, False Negtive], [False Postive, Negtive]) corresponding to [infected, non-infected] individulas
    for i in range(no_individuals):
        if people_sample[i] :
            test_results[i] = np.random.choice([True, False], p=[0.999, 0.001])
        else:
            test_results[i] = np.random.choice([True, False], p=[0.005, 0.995])
    # Calcaulte probability that the test returns a positive result from medical test result
    all_positive_tests = np.sum(test_results)
    # Calcaulte probability that the test returns a true positive result for person has contracted the virus
    true_positives = np.sum(people_sample & test_results)
    #Calcaulte the probability that the person has the virus given medical test is positve
    prob_virus_given_positive = true_positives / all_positive_tests
    
    return prob_virus_given_positive
        

N = 10**4
num_simulations = 1000
simulation_results = [virus_test(N) for _ in range(num_simulations)]
print(simulation_results[0])

# Plot results
plt.hist(simulation_results, bins=50, edgecolor='black')
plt.title('Distribution of P(V|Tp) across simulations')
plt.xlabel('P(V|Tp)')
plt.ylabel('Frequency')
plt.axvline(x=0.80305, color='r', linestyle='dashed', linewidth=2, label='Theoretical value')
plt.legend()
plt.show()
```

### Answers: 
#### Question 1
Let  ð‘ˆ  be the event describing that the test for a pooled sample of k people will be positive.
##### 1. Calculate Probability(Pooled sample test positive)
If one person of the group is positive, then the whole group will be positive, which means: \
To have a negative test, everyone of the group has to be negative. Thus

$$
P(\bar{U}) = (1-p)^k
$$

then

$$
P(U) = 1 - P(\bar{U}) = 1-(1-p)^k
$$

##### 2. Calculate Expection(Number of tests necessary)
According to the topic, we can conclude the following relationship:

$$
\begin{aligned}P(\text{test num = 1} )&=P(U)=(1-P)^k\\P(\text{test num = k+1})&=P(U)=1-(1-P)^k\end{aligned}
$$

Thus for one group:

$$
\begin{aligned}E(\text{testsÂ forÂ oneÂ group})&=\sum{x\cdot p(x)}\\&=\left[1-\left(1-p\right)^{k}\right]\cdot\left(k+1\right)+\left(1-p\right)^{k}\cdot1\\&=k+1-k\left(1-p\right)^{k}\end{aligned}
$$

Given that

$$
\text{Group number} = \frac{n}{k}
$$

For total groups:

$$
E(\text{total tests})=E(\text{testsÂ forÂ oneÂ group})\times \text{Group number}=\frac{n}{k}\times\left(k+1-k(1-p)^k\right)
$$


```python
# Placeholder for python part of exersice 15
# Question 1
import numpy as np

#set up
n = 10**2 # Number of people
k = 10    # Number of groups
no_iteration = 10**4
p = [0.02, 0.2, 0.6]
# calculate Theoretical and Empirical
e_th = np.zeros(3)  # for the Expecation of necessary tests
p_th = np.zeros(3)  # for the Probability of pooled sample of k people will be positive
for i in p:
    #Theoretical
    p_th = 1 - (1 - i)**k
    e_th = (n / k) * (1 * (1 - i)**k + ((k + 1) * (1 - (1 - i)**k)))
    # Empirical
    pooled_results = []
    for _ in range(no_iteration):
        individual_results = np.random.binomial(1, i, size=n)
        pooled_result = np.any(individual_results.reshape(-1, k), axis=1)
        pooled_results.append(pooled_result.mean())
    print(f"When p = {str(i)} , Theoretical expected number of tests = {e_th:.4f} the empirical expected number of tests: {(n / k)*(1 * (1 - np.mean(pooled_results)) + (1 + k) * np.mean(pooled_results)):.4f}\n")
```

#### Question 2
Let 

$$
    f(p)=(1-p)^k
$$

According the the Taylor series expansion:

$$
    f(p)=f(a)+f'(a)(p-a)+\frac{f''(a)}{2!}(p-a)^2+\frac{f'''(a)}{3!}(x-p)^3+\ldots 
$$

Expand $f(p)$ around $p=0$, we can approximate:

$$
    f(p)\approx f(0)+f'(0)(p-0)=1-kp
$$

Substituting this approximation into the expression for the expected total number of tests, and regard $E$ is a function of $k$:

$$
    E(k)\approx\frac{n}{k}\times(k+1-k(1-kp))
$$

Simplifying:

$$
    E(k)\approx n\left(\frac1k+kp\right)
$$

Take the derivative of $E(k)$ with respect to $k$ and set it equal to zero:

$$
    \frac{d}{dk}\left(n\left(\frac{1}{k}+kp\right)\right)=n\left(-\frac{1}{k^2}+p\right)=0
$$

Simplifying:

$$
    k=\frac1{\sqrt{p}}
$$

##### Conclusion
If assuming $p$ is close to $0$, the best value of $k$ is approximated to $k=\frac1{\sqrt{p}}$

```python
# python part for exercise 15(question 2)
import numpy as np

# set up
p_list = [0.02, 0.2, 0.6]
k_list = range(1, 100)  # num of 'k'

e = np.zeros((len(k_list), len(p_list)))


def e_test(p, k):
    return k + 1 - k * (1 - p ** k)


# find the best k
for p_m in range(len(p_list)):
    for k_m in k_list:
        e[k_m-1, p_m] = e_test(p_list[p_m], k_m)

best_k = np.argmin(e, axis=0) + 1

# Approximationa
k_app = np.zeros(3)
for i_m in range(len(p_list)):
    k_app[i_m] = 1 // (p_list[i_m])**0.5

for i in range(len(p_list)):
    print('When p = ' + str(p_list[i]) + ',theoretical best k = ' + str(best_k[i]) + ', approximated k = ' + str(k_app[i]))
```

##### Conclusion
As the result shows, when $p$ is close to $0$(like $p=0.02$), the The approximate results are close to the theoretical results.

While as $p$ increases, the difference between approximate results and theoretical results are bigger.

#### Question 3

For which values of p is pooling better than just testing every individual?

- group testing is better than testing every individual when $\text{E}[Necessary Tests]$ using the pooling method is less than n (the number of people).
- For small p, the pooling strategy is more efficient as long as the positive test probability is low.
- Using the approximation for small p, we have:
$\text{E}[Necessary Tests]\approx n\left(\frac1k+kp\right)$

- This is less than n when:
$n\left(\frac1k+kp\right) < n$
To find the optimal value of k we minimize the left-hand side of this inequality.
This requires solving for k that minimizes the function:

$$
    f(k)= \frac{1}{k} + kp
$$

$$
    f'(k) =  -\frac{1}{k^2} + p
$$

Setting the derivative equal to 0:

$$
-\frac{1}{k^2} + p = 0
$$

$$
\frac{1}{k^2} = p
$$

$$
    k=\frac1{\sqrt{p}}
$$


Substituting this into the expression for $E(k)$:

$$
    E(k)\approx n\left(\frac{1}{\frac{1}{\sqrt{p}}}+\frac{1}{\sqrt{p}}p\right)
$$

Simplifying:

$$
    E(k)\approx n\times2\sqrt{p}
$$

Thus when pooling is better than just testing every individual, following inequation should be satisfied:

$$
    2n\sqrt{p}<n
$$

Simplifying:

$$
    p<\frac{1}{4}    
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# set up
n = 10**2
p_list = np.linspace(0.01, 0.5, 500)

e_indi = np.full((len(p_list), ), n)
e_th = np.zeros(len(p_list))
e_app = np.zeros(len(p_list))


# calculate e_theoretical & e_approximate
for p_m in range(len(p_list)):

    optimal_k = int(1 / np.sqrt(p_list[p_m]))
    if optimal_k > 99:
        optimal_k = 99

    e_th[p_m] = n / optimal_k * (optimal_k + 1 - optimal_k * (1 - p_list[p_m])**optimal_k)
    e_app[p_m] = 2 * n * np.sqrt(p_list[p_m])


# plot
plt.plot(p_list, e_th, label='Theoretical Expected Test Number(Pooling)', color='blue')
plt.plot(p_list, e_app, label='Approximated Expected Test Number(Pooling)', color='red')
plt.plot(p_list, e_indi, label='Number of Individual Tests', color='green')
plt.xlabel('Probability of Positive Result')
plt.ylabel('Expected Number of Tests')
plt.title('Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

##### Conclusion
The chart inllustrates that:

1. The comparison between pooled test and individual test 
    - When $0.01\leq p\leq0.4$
        - Pooled test is better than individual test.
    - When $0.4\leq p\leq0.5$
        - Individual test is better.
        
2. The difference between theoretical expectation and approximated expectation
    - when $p$ is close to $0$, they are close.
    - The difference become larger as $p$ is bigger.
